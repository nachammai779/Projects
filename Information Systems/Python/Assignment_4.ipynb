```python
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["Bitcoin","Artificial Intelligence"]
```

```python
pytrends.build_payload(kw_list, cat=0, timeframe='2019-07-01 2019-07-30', geo='', gprop='')
```

```python
data=pytrends.interest_over_time()
```

```python
data
```

```python
data[["Bitcoin","Artificial Intelligence"]].sum()
```

```python
plot_m=data[["Bitcoin","Artificial Intelligence"]].plot()
```
